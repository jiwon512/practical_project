# 유지방 유단백 NaN값인 애들 조회

```python
# 유지방 또는 유단백 결측인 행만 추출
missing_fat_protein = df[df['유지방'].isna() | df['유단백'].isna()]
missing_fat_protein.head()
```

```python
missing_object = missing_fat_protein[['농장아이디','개체번호']].dropna().drop_duplicates().sort_values(by='농장아이디', ascending=True)
missing_object
```

# 시간대별 결측 비율 확인

```python
# 시간 변환
df['착유시작일시'] = pd.to_datetime(df['착유시작일시'], errors='coerce')

target_all_data = df[df['개체번호'].isin(target_ids)].copy()

# 결측 여부 플래그
target_all_data['유성분_결측'] = target_all_data[['유지방', '유단백']].isna().any(axis=1)
target_all_data['시각'] = target_all_data['착유시작일시'].dt.hour
```

```python
# 시간대별 결측 비율
target_all_data['시각'] = target_all_data['착유시작일시'].dt.hour
hourly_missing = target_all_data.groupby('시각')['유성분_결측'].mean()
```

시각	결측률	해석
0시 -	0.000	0시에 유성분 결측 없음
5시	- 0.230	23%가 결측 — 상대적으로 높음
14시	- 0.107	14시(오후 2시경)도 일부 존재
18시	- 0.133	18시(저녁 6시경)에 결측률 상승
8~12시	- 대부분 0.0	오전은 안정적

```python
hourly_missing
```

```python
# 3. 전체 데이터 중 해당 개체들의 모든 이력 추출
target_ids = missing_object['개체번호'].unique()
target_all_data = df[df['개체번호'].isin(target_ids)].copy()

# 4. 유성분 결측 여부 플래그 생성
target_all_data['유성분_결측'] = target_all_data[['유지방', '유단백']].isna().any(axis=1)

# 5. 센서값 평균 비교
sensor_cols = ['전도도', '온도', '공기흐름']  # 혈액흐름은 문자형이라 제외
sensor_comparison = target_all_data.groupby('유성분_결측')[sensor_cols].mean()
```

```python
sensor_comparison
#전도도는 큰 차이 없음
#온도, 공기흐름이 결측 시 뚜렷하게 낮음(센서 미작동/기기 오류 시 결측 발생 가능성)
```

# 착유량 값 별 빈도수 계산 및 이상치 탐지

```python
# 착유량 값별 빈도수 계산 후 데이터프레임으로 정리
milk_yield_dist = (
    target_all_data['착유량']
    .value_counts()
    .sort_index()
    .reset_index()
)

# 컬럼명 정리
milk_yield_dist.columns = ['착유량', '빈도']
```
