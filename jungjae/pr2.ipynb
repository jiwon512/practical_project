# 파일 살펴보기

```python
# 시간 변환
df['착유시작일시'] = pd.to_datetime(df['착유시작일시'], errors='coerce')
```

```python
df
```

```python
df = df[df['농장아이디'] != '20261']
```

```python
# 혈액흐름 값별 빈도수 계산 후 데이터프레임으로 정리
blood = (
    df['혈액흐름']
    .value_counts()
    .sort_index()
    .reset_index()
)
# 컬럼명 정리
blood.columns = ['혈액흐름', '빈도']
blood
```

```python
farm_obj_map = (
    df.groupby('농장아이디')['개체번호']
    .nunique()
    .reset_index()
)
# 컬럼명 정리
farm_obj_map.columns = ['농장아이디', '개체수']

farm_obj_map
```

```python
# 축종코드 값별 빈도수 계산 후 데이터프레임으로 정리
ccode = (
    df['축종코드']
    .value_counts()
    .sort_index()
    .reset_index()
)
# 컬럼명 정리
ccode.columns = ['축종코드', '빈도']
ccode
```

```python
#착유회차
milk_c = (
    df['착유회차']
    .value_counts()
    .sort_index()
    .reset_index()
)
# 컬럼명 정리
milk_c.columns = ['착유회차', '빈도수']

milk_c
```

```python
#전도도
conductivity = (
    df['전도도']
    .value_counts()
    .sort_index()
    .reset_index()
)
# 컬럼명 정리
conductivity.columns = ['전도도', '빈도수']
conductivity
```

```python
#온도
temp = (
    df['온도']
    .value_counts()
    .sort_index()
    .reset_index()
)
# 컬럼명 정리
temp.columns = ['온도', '빈도수']
temp
```
