```python
# 시간 변환
df['착유시작일시'] = pd.to_datetime(df['착유시작일시'], errors='coerce')
# 시간 변환
df['착유종료일시'] = pd.to_datetime(df['착유종료일시'], errors='coerce')
```

```python
# 1. 제거할 컬럼 리스트
drop_cols = ['축종코드', '제조사 아이디', '측정일시', '수집일자', '수집건수','체세포']

# 2. 컬럼 제거 + 농장아이디 20261인 행 제거
df = df.drop(columns=drop_cols)
df = df[df['농장아이디'] != 20261]
```

# 이상치 탐지(착유시간)

```python
# 착유 시간(분) 계산
df['착유시간(분)'] = (df['착유종료일시'] - df['착유시작일시']).dt.total_seconds() / 60
```

```python
df
```

```python
# 이상치 여부 (IQR 방식)
def iqr_outlier_check(x):
    q1 = x.quantile(0.25)
    q3 = x.quantile(0.75)
    
    iqr = q3 - q1

    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return ~x.between(lower, upper)  # 이상치인 경우 True

df['이상치여부'] = df['착유시간(분)'].transform(iqr_outlier_check)
df
```

```python
#이상치 여부
trouble = (
    df['이상치여부']
    .value_counts()
    .sort_index()
    .reset_index()
)
trouble.columns = ['이상치여부', '갯수']
trouble
```

```python
# 2. 이상치만 필터링한 DataFrame
outliers_df = df[df['이상치여부'] == True]
outliers_df
```
